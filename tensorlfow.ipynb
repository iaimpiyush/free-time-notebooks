# Digit Recognition with TensorFlow

This notebook demonstrates how to build a simple Neural Network to recognize handwritten digits (0-9) using the famous **MNIST dataset**. 

### Steps involved:
1. **Importing Libraries**: Loading the necessary tools.
2. **Loading Data**: Getting the images of handwritten numbers.
3. **Preprocessing**: Preparing the images for the computer to read.
4. **Building the Model**: Designing the "brain" of the AI.
5. **Compiling**: Setting the rules for learning.
6. **Training**: Let the computer practice and learn.
7. **Evaluation**: Testing how well the computer learned.

### Step 1: Import Libraries
We need TensorFlow and Keras to build the neural network, and the MNIST dataset is built right into Keras.

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
import tensorflow as tf
```

### Step 2: Load the Dataset
The dataset is split into `train` (for learning) and `test` (for checking accuracy later).

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

### Step 3: Preprocessing the Data
We "flatten" the 2D square images (28x28) into 1D lines (784 pixels) and change pixel values from 0-255 to 0-1 (Normalizing) to help the model learn faster.

```python
train_images = train_images.reshape(
    (train_images.shape[0], 28*28)).astype('float32') / 255
test_images = test_images.reshape(
    (test_images.shape[0], 28*28)).astype('float32') / 255
```

### Step 4: Building the Neural Network
We use a `Sequential` model with two layers:
- A **Dense** layer with 128 neurons (the "hidden" layer).
- A **Dense** layer with 10 neurons (one for each digit 0-9).

```python
model = Sequential([
    Dense(128, activation='relu', input_shape=(28 * 28,)),
    Dense(10, activation='softmax')
])
```

### Step 5: Compiling the Model
Here we define how the model measures its performance (`loss`) and how it improves (`optimizer`).

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

### Step 6: Training the Model
The computer will look at the images 5 times (`epochs`) to learn the patterns.

```python
model.fit(train_images, train_labels, epochs=5)
```

### Step 7: Evaluating the Results
Finally, we test the model on the "test dataset" to see how accurate it really is.

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
```

```python
print(f"Test Accuracy: {(test_acc)*100}")
print(f"Test Loss: {(test_loss)*100}")
```
